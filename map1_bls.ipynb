```python
import folium
import pandas as pd
import random
```

```python
bls = pd.read_csv("bls.csv")
m = folium.Map(location=[pd.Series.mean(bls['LAT']), pd.Series.mean(bls['LON'])])
```

```python
colors = [
    'red',
    'blue',
    'gray',
    'darkred',
    'lightred',
    'orange',
    'beige',
    'green',
    'darkgreen',
    'lightgreen',
    'darkblue',
    'lightblue',
    'purple',
    'darkpurple',
    'pink',
    'cadetblue',
    'lightgray',
    'black'
]

fg = folium.FeatureGroup(name='my map features')
    
```

```python
for i in range(len(bls.index)):
    fg.add_child(folium.Marker(location= [bls["LAT"][i], bls["LON"][i]], popup=bls['PLACE'][i], icon= folium.Icon(color=random.choice(colors))))
```

```python
m.add_child(fg)
```
